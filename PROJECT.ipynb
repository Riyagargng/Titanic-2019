```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            return 37
        elif Pclass ==2:
            return 29
        elif Pclass ==3:
            return 24
    else:
        return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(impute_age,axis =1)
```

```python
sns.heatmap(train.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
train.drop('Cabin',axis = 1,inplace = True)
```

```python
sns.heatmap(train.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
train.dropna(inplace=True)
```

```python
train.isnull().sum()
```

```python
def impute_sex(cols):
    Sex=cols[0]
    if Sex == 'male':
            return 0
    elif Sex== 'female':
            return 1
```

```python
train['Sex']=train[['Sex']].apply(impute_sex,axis=1)
```

```python
train
```

```python
def impute_embarked(cols):
    Embarked=cols[0]
    if Embarked == 'S':
            return 0
    elif Embarked=='C':
            return 1
    elif Embarked =='Q':
            return 2
```

```python
train['Embarked']=train[['Embarked']].apply(impute_embarked,axis=1)
```

```python
train
```

```python
train.drop(['Name','Ticket','PassengerId'],axis =1,inplace=True)
```

```python
train
```

```python
x = train.drop('Survived',axis=1)#will select all the colums except the survived 
y = train['Survived']#will select the named column
```

```python
plt.plot(x,y,color='r')
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3)
```

```python
reg.fit(x_train,y_train)
```

```python
a=reg.predict(x_test)
a
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,a)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,a))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,a)
```
